```python
def reduce_other(row, genre_mapping):
    if row['genres_0'] != 'Other':
        return row['genres_0']
    for col in ['genre_2', 'genre_3', 'genre_4', 'genre_5']:
        if pd.notna(row[col]) and row[col] in genre_mapping:
            return genre_mapping[row[col]]
    return 'Other'

# Apply the function to update the 'genres_0' column
df['genres_0'] = df.apply(reduce_other, axis=1, genre_mapping=genre_mapping)
```

```python
# Function to calculate the percentage of books with worse rating within the same genre
def calculate_worse_rating_percentage(row, genre_group):
    total_books = genre_group.shape[0]
    worse_books = genre_group[genre_group['rating'] < row['rating']].shape[0]
    return (worse_books / total_books) * 100

# Group the dataframe by 'genres_0'
df['worse_rating_percentage'] = df.groupby('genres_0').apply(lambda group: group.apply(lambda row: calculate_worse_rating_percentage(row, group), axis=1)).values
```

```python
# Sort the dataframe by 'genres_0' and 'rating'
dc = df.sort_values(by=['genres_0', 'rating'])

# Group by 'genres_0' and calculate the total books and cumulative count of worse ratings
dc['total_books'] = dc.groupby('genres_0')['rating'].transform('count')
dc['cumulative_worse_ratings'] = dc.groupby('genres_0').cumcount()

# Calculate the percentage of books with worse ratings
dc['worse_rating_percentage'] = (dc['cumulative_worse_ratings'] / dc['total_books']) * 100

# Drop the helper columns
dc = dc.drop(columns=['total_books', 'cumulative_worse_ratings'])
```
