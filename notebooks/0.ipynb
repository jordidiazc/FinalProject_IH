```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/thege/OneDrive/Escritorio/IH/FINAL/data/raw/GoodReads_100k_books.csv")
df.head()
```

```python
df=df.dropna(subset=['genre']) 
```

```python
df=df.dropna(subset=['desc']) 
```

```python
df
```

```python
df_separado = df['genre'].str.split(',', expand=True)
df_separado
```

```python
df_separado.columns = [f'genre_{i+1}' for i in range(df_separado.shape[1])]
```

```python
df_separado
```

```python
df = pd.concat([df, df_separado], axis=1)
```

```python
df
```

```python
import matplotlib.pyplot as plt

bins = [-1, 0.5, 199, 399, 599, 799, float('inf')]
labels = ['0', '1-199', '200-399', '400-599', '600-799', '800+']
df['pages_binned'] = pd.cut(df['pages'], bins=bins, labels=labels, right=False)

# Crear el diagrama de barras usando la columna 'pages_binned'
df['pages_binned'].value_counts().sort_index().plot(kind='bar')

# Añadir etiquetas y título al gráfico
plt.xlabel('Intervalo de páginas')
plt.ylabel('Frecuencia')
plt.title('Distribución de la columna "pages" en intervalos específicos')

# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```
