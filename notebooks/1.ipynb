```python
columns_to_drop = ['isbn13','genre_6','genre_7','genre_8','genre_9','genre_10','genre_11','genre_12','genre_13','genre_14','genre_15','genre_16','genre_17','genre_18','genre_19','genre_20']
df=df.drop(columns=columns_to_drop)
```

```python
bins = [-1, 0.5, 299, 599,float('inf')]
labels = ['0', '-300', '300-599','600+']
df['pages_binned_300'] = pd.cut(df['pages'], bins=bins, labels=labels, right=False)
```

```python
df['bookformat'] = df['bookformat'].apply(lambda x: x if x in ['Paperback', 'Hardcover'] else np.nan)
```
